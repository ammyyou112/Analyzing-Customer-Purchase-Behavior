```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Load dataset
df = pd.read_excel('online_retail_II.xlsx')
df
```

```python
#Check Duplications
duplicate_rows = df[df.duplicated()]
duplicate_rows
```

```python
#Remove duplications
df = df.drop_duplicates()
df
```

```python
#Check missing values
missing_values = df.isna()
missing_counts = missing_values.sum()
missing_counts
```

```python
print(df)
```

```python
#Drop rows of null values from customer id
df_copy = df.copy()
df_copy.dropna(subset = ['Customer ID'], inplace = True)
df_copy
```

```python
#Fill null values from description column
description_counts = df_copy['Description'].value_counts()
most_frequent_des = description_counts.index[0]
df_copy['Description'].fillna(most_frequent_des, inplace=True) #Sheet1
df_copy
 
```

```python
#Check is there any wrong value in dataset or not
df_copy['Quantity'] = df_copy['Quantity'].astype(str)
price_string = df_copy['Quantity'].str.contains(r'^\$\d+\.\d{2}$')
price_string.sum()
```

```python
df_copy.describe()
```

```python
#Identify the top 10 products 
top10_products = df_copy['Description'].value_counts().head(10)
top10_products.plot(kind='bar')
plt.title('Top 10 Products')
plt.xlabel('Product Description')
plt.ylabel('Number of Purchases')
plt.show()
```

```python
#Analyze the top sales in countries
sales_by_country = df_copy.groupby(['Description', 'InvoiceDate'])['Country'].agg(['count'])
sales_by_country = sales_by_country.sort_values(by='count', ascending=False)

top_countries = sales_by_country.head(10)
plt.bar(x=[str(x) for x in top_countries.index], height=top_countries['count'])

plt.title('Top 10 Sales Countries')
plt.xlabel('Countries')
plt.ylabel('Sales Count')

plt.xticks(rotation=90)

plt.show()
```

```python
#Identify the top customers by customer total spent
total_customers = df_copy.groupby('Customer ID')['Price'].sum()
total_customers = total_customers.sort_values(ascending=False)
top_customers = total_customers.head(10)

plt.bar(x=[str(x) for x in top_customers.index], height=top_customers.values)
plt.title('Top 10 Customers by Total Spent')
plt.xlabel('Customer ID')
plt.ylabel('Total Spent')
plt.xticks(rotation=90)
plt.show()

```
